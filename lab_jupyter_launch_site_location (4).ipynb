```python
## Task 1: Mark all launch sites on a map
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. 
# In addition, add Launch site name as a popup label
for index, site in launch_sites_df.iterrows():
    coordinate = [site['Lat'], site['Long']]
    site_name = site['Launch Site']
    
    # Create Circle
    circle = folium.Circle(
        coordinate, 
        radius=1000, 
        color='#000000', 
        fill=True
    ).add_child(folium.Popup(site_name))
    
    # Create Marker (using DivIcon for the text label)
    marker = folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style=\"font-size: 12; color:#d35400;\"><b>%s</b></div>' % site_name,
        )
    )
    
    site_map.add_child(circle)
    site_map.add_child(marker)

site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Create a function to assign marker color based on the 'class' value
def assign_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

# Apply the function to create the new column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Display the tail to verify the new column
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate and customize the icon property with the marker_color
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate and color
    marker = folium.Marker(
        [record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    # Add marker to the cluster
    marker_cluster.add_child(marker)

site_map
```

```python
from math import radians, cos, sin, asin, sqrt
import folium
from folium.features import DivIcon

# Define distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points on the Earth (in kilometers).
    Uses the Haversine formula.
    """
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Earth's radius in kilometers
    return c * r

# --- Your existing task code ---
launch_site_df_ksc = launch_sites_df[launch_sites_df['Launch Site'] == 'KSC LC-39A'].iloc[0]
launch_site_lat = launch_site_df_ksc['Lat']
launch_site_lon = launch_site_df_ksc['Long']

# Coordinate of the closest coastline
coastline_lat = 28.56294  
coastline_lon = -80.56781 
coastline_coordinate = [coastline_lat, coastline_lon]
launch_site_coordinate = [launch_site_lat, launch_site_lon]

# Calculate the distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to closest coastline: {distance_coastline:.2f} KM")

# Create and add a folium.Marker
distance_marker = folium.Marker(
   coastline_coordinate,
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_coastline:.2f} KM</b></div>',
   )
)
site_map.add_child(distance_marker)

# Create a line between launch site and coastline
coordinates = [launch_site_coordinate, coastline_coordinate]
lines = folium.PolyLine(locations=coordinates, weight=1, color='orange')
site_map.add_child(lines)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
#  Create and add a folium.Marker on the selected closest coastline point
#  Display the distance between coastline point and launch site using the icon property

import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# --- Coordinates ---
# Replace these with your actual points
launch_site_lat, launch_site_lon = 28.562302, -80.577356
coastline_lat, coastline_lon = 28.56367, -80.57163

# --- Calculate distance (Haversine formula) ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius (km)
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# --- Create map ---
m = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# --- Launch site marker ---
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color="green", icon="rocket", prefix="fa")
).add_to(m)

# --- Coastline marker ---
folium.Marker(
    [coastline_lat, coastline_lon],
    popup="Closest Coastline",
    icon=folium.Icon(color="blue", icon="flag", prefix="fa")
).add_to(m)

# --- Add line connecting both points ---
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color="red", weight=2.5, opacity=0.8
).add_to(m)

# --- Add distance label marker ---
mid_lat = (launch_site_lat + coastline_lat) / 2
mid_lon = (launch_site_lon + coastline_lon) / 2

distance_marker = folium.Marker(
    location=[mid_lat, mid_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance)
    )
)
distance_marker.add_to(m)

print(f" Distance between launch site and coastline: {distance:.2f} km")
m
```
